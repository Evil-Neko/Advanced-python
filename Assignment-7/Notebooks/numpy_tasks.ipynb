```python
import numpy as np
arr1 = np.arange(1, 11)
print("Arr1:", arr1)
arr2 = np.random.randint(1, 10, size=(3, 3))
print("Arr2:",arr2)

sum_arr1 = arr1.sum()
mean_arr2 = arr2.mean()
max_arr2 = arr2.max()
print(f"Sum: {sum_arr1}, Mean: {mean_arr2}, Max: {max_arr2}")

arr1_doubled = arr1 * 2
print("Arr1 all mult by 2:", arr1_doubled)

arr2_5is0 = arr1.copy()
arr2_5is0[arr2_5is0 > 5] = 0
print("Arr2 but all 5s are 0s", arr2_5is0)
```
