```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Name": ["Dinmukhamed", "Imran", "Suleiman", "Miroslav", "Dair"],
    "Age": [21, 20, 20, 20, 21],
    "Score": [85, 80, 90, 88, 95]
}

df = pd.DataFrame(data)
plt.figure()
plt.plot(df["Name"], df["Score"])
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
plt.figure()
plt.bar(df["Name"], df["Score"])
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Bar Chart")
plt.show()
```
